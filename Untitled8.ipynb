```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("/content/penguins_size.csv")
data.head()
```

```python
data.info()
```

```python
#check for null values
data.isnull().sum()
```

```python
#drop the null values
data.dropna(inplace=True)
```

```python
#Plot a graph of gender
sns.countplot(x='sex',data=data,palette="Dark2")
plt.title("Gender distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

#Findings
* There is no big difference between the number of male penguins and fealmale penguins.

```python
#Plot a graph of gender
sns.countplot(x='species',data=data,palette="Dark2")
plt.title("Species distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

#Findings
* The adelie species has the largest number of species followed by gentle and chinstrap being the lowest.

```python
#Plot a graph of island
sns.countplot(x='island',data=data,palette="Dark2")
plt.title("Penguin distribution per Island")
plt.xlabel("island")
plt.ylabel("count")
plt.show()
```

#Findings
* The biscoe island has the largest number of pengunis then dream and torgersen having the least.

```python
#Plot a graph of island
sns.countplot(x='island',hue="species",data=data,palette="Dark2")
plt.title("Penguin distribution per island")
plt.xlabel("island")
plt.ylabel("Count")
plt.show()
```

#Findings
*Torgersen only has Adelie species.
*Biscoe has both Adelie and chinstrap.
*Deram has both Gentoo and Adelie.
*Adelie is on every island.
*Chinstrap is only found on dream island.
